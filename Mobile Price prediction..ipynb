# Importing a Machine learning Algorithm

## We choose a lianear Regression for our dataset

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## Training a data using train_test_split function

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(['price_range'],axis=1), 
                                                    data['price_range'], test_size=0.20, 
                                                    random_state=8)
```

## Fiting a data in linear Regression (model) using a fit function

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

# Accuracy of the model

```python
accuracy = model.score(X_test, y_test)
print(accuracy*100,'%')
```

# prediction of the model

```python
predictions
```
